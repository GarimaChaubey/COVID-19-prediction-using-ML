```python
import pandas as pd
```

```python
covid=pd.read_csv('covid_19_india.csv')
```

```python
covid.info()
```

```python
covid.describe()
```

```python
covid.head()
```

```python
covid=covid[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
```

```python
covid.head()
```

```python
covid=covid.rename(columns={'Date':'date', 'State/UnionTerritory':'state', 'Cured':'cured','Deaths':'deaths', 'Confirmed':'confirmed'})
```

```python
covid.head()
```

```python
covid.tail()
```

```python
covid.shape
```

```python
latest=covid[covid['date']=='2021-06-24']
```

```python
latest
```

```python
latest.sort_values(by='confirmed', ascending=False)
```

```python
top_5_states=latest.sort_values(by='confirmed', ascending=False).head()
```

```python
top_5_states
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x="state", y="confirmed", data=top_5_states)
plt.show()
```

```python
bottom_5_states=latest.sort_values(by="confirmed", ascending=True).head()
```

```python
bottom_5_states
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x="state", y="confirmed",data=bottom_5_states)
```
